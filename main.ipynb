# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from Data.DataGenerator import DataGenerator
from Essentials.CleverPolynomialRegression import CleverPolynomialRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
# Muting warnings
import warnings
warnings.filterwarnings("ignore")
```

# 2 Dimencional Data

```python
# Reading data
data_2d = pd.read_csv('Data/dataset0.csv')
data_2d.columns = ['x', 'y']
print('Number of rows:', len(data_2d))
data_2d.sample(10)
```

```python
# Data split
data_2d = data_2d.to_numpy()
X_2d, y_2d = data_2d[::, 0], data_2d[::, 1]
```

```python
# Plot
ax = plt.axes()
ax.scatter(X_2d, y_2d, c=y_2d, s=5)
plt.show()
```

```python
# Data scale
scaler_2d = StandardScaler()
X_2d_scalled = scaler_2d.fit_transform(X_2d.reshape(-1, 1))

# Data split
X_2d_train, X_2d_test, y_2d_train, y_2d_test = train_test_split(X_2d_scalled, y_2d, test_size=0.2, random_state=1)
```

```python
# Parameters
alpha_2d: int = 1
divisions_2d: int = 3
power_decider_parameter_2d: str = 'mean'
model_2d: LinearRegression = LinearRegression(n_jobs=-1)

# Model
cpr_2d: CleverPolynomialRegression = CleverPolynomialRegression(alpha_2d, divisions_2d, power_decider_parameter_2d, model_2d)

# Fitting and scoring
powers_2d = cpr_2d.fit(X_2d_train, y_2d_train)
cpr_2d.score(X_2d_test, y_2d_test)
print('Powers:', powers_2d)
```

# 3 Dimencional Data

```python
# Reading data
data_3d = pd.read_csv('Data/dataset1.csv')
data_3d.columns = ['x', 'y', 'z']
print('Number of rows:', len(data_3d))
data_3d.sample(10)
```

```python
# Data split
data_3d = data_3d.to_numpy()
X_3d, y_3d = data_3d[::, :-1], data_3d[::, -1]
```

```python
# Plot
ax = plt.axes(projection='3d')
ax.scatter(X_3d[::, 0], X_3d[::, 1], y_3d, c=y_3d, s=5)
plt.show()
```

```python
# Data scaling
scaler_3d = StandardScaler()
X_3d_scalled = scaler_3d.fit_transform(X_3d)

# Data split
X_3d_train, X_3d_test, y_3d_train, y_3d_test = train_test_split(X_3d_scalled, y_3d, test_size=0.2, random_state=1)
```

```python
# Parameters
alpha_3d: int = 1
divisions_3d: int = 2
power_decider_parameter_3d: str = 'median'
model_3d: LinearRegression = LinearRegression(n_jobs=-1)

# Model
cpr_3d: CleverPolynomialRegression = CleverPolynomialRegression(alpha_3d, divisions_3d, power_decider_parameter_3d, model_3d)

# Fitting and scoring
powers_3d = cpr_3d.fit(X_3d_train, y_3d_train)
cpr_3d.score(X_3d_test, y_3d_test)
print('Powers:', powers_3d)
```
